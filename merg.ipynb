```python
import pandas as pd
```

```python
df = pd.read_csv('my_submission_2.csv')
df
```

```python
df_rez = pd.read_csv('my_submission_2_with_nan.csv', index_col=0)
df_rez
```

```python
df_merge = pd.merge(df, df_rez, on='id', how='left')
df_merge
```

```python
df_merge.per_square_meter_price_y[df_merge.per_square_meter_price_y.isna()] = df_merge.per_square_meter_price_x[df_merge.per_square_meter_price_y.isna()]
df_merge
```

```python
df_merge.rename(columns={'per_square_meter_price_y':'per_square_meter_price'}, inplace=True)
df_merge.drop(columns='per_square_meter_price_x', inplace=True)
df_merge
```

```python
df_merge.to_csv('out_2.csv')
```
